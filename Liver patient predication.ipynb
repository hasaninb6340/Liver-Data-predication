# importing libraries and data understanding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Liver patient.csv")
dataset.head()
```

```python
dataset.dtypes
```

```python
dataset.isnull().sum()
```

```python
dataset["Albumin_and_Globulin_Ratio"].describe()
```

```python
dataset["Albumin_and_Globulin_Ratio"]=dataset["Albumin_and_Globulin_Ratio"].fillna(dataset["Albumin_and_Globulin_Ratio"].mean())
```

```python
dataset.isnull().sum()
```

```python
import seaborn as sns
fig=sns.pairplot(dataset, hue="Dataset")
fig.savefig("pairplot.png")
```

```python
dataset.corr()
```

```python
dataset["Bilirubinratio"]=dataset["Direct_Bilirubin"]/dataset["Total_Bilirubin"]
```

```python
df=pd.get_dummies(dataset["Gender"])
df
```

```python
dataset=pd.concat([dataset,df],axis=1)
```

```python
dataset
```

```python
dataset=dataset.drop(["Gender"],axis=1)
```

```python
dataset=dataset.drop(["Total_Bilirubin"],axis=1)
dataset=dataset.drop(["Direct_Bilirubin"],axis=1)
```

Splitting data into X and y data set

```python
X=dataset.iloc[:,[0,1,2,3,4,5,6,8,9,10]]
```

```python
y=dataset.iloc[:,7]
y
```

# Scaling and splitting data into test and train

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn import preprocessing
train_scaled = preprocessing.scale(x_train)
test_scaled=preprocessing.scale(x_test)
```

```python
test_scaled
```

# Knn algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(train_scaled,y_train)
pre=classifier.predict(test_scaled)
pre
```

Evaluating knn

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
print(confusion_matrix(y_test, pre))
print(f1_score(y_test,pre))
pre=classifier.predict(train_scaled)
print(confusion_matrix(y_train, pre))
print(f1_score(y_train,pre))
```

```python
error = []

# Calculating error for K values between 1 and 40
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_scaled, y_train)
    pred_i = knn.predict(test_scaled)
    error.append(np.mean(pred_i != y_test))
min = error[0];    
result=0     
#Loop through the array    
for i in range(0, len(error)):    
    #Compare elements of array with min    
    if(error[i] < min):
        min=error[i]
        result=i
plt.figure(figsize=(12, 6))
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error') 
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=26)
classifier.fit(train_scaled,y_train)
pre=classifier.predict(test_scaled)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
print(confusion_matrix(y_test, pre))
print(f1_score(y_test,pre))
pre=classifier.predict(train_scaled)
print(confusion_matrix(y_train, pre))
print(f1_score(y_train,pre))
```
